### Eigen Value Decomposition for square matrix - Maths behind the concept 
#### Used - Google translator

```python
import numpy as np
```

Square symmetric matrix - Eigen Value Decomposition

```python
R = np.random.randn(10,5)
```

```python
A = np.matmul(R,R.T)
```

```python
A
```

```python
A.shape
```

### Apply EVD

```python
evd = np.linalg.svd(A)
```

type(evd) == tuple

#### A = Q.lambda.Q^T
Each matrix is 10X10 matrix

```python
evd
```

```python
evd[0].shape
```

#### Calculating the L2 norm of the evd

```python
np.linalg.norm(evd[0],ord=2,axis=0)
```

```python
evd[1]
```

This the eigen value of the matrix

```python
q1_transpose = evd[2][0,:]
```

```python
q1_transpose
```

```python
q1_transpose = q1_transpose.reshape(1,q1_transpose.shape[0])
```

```python
q1_transpose
```

### Dot product of q1_transpose with q1

```python
np.matmul(q1_transpose,evd[0])
```

### Calcualting the rank of the matrix
